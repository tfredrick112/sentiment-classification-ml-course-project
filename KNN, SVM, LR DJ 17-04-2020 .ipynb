```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval
```

```python
df= pd.read_csv('./ml-course-project-master/processed_trainingset.csv')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#index=np.where(df['processed_tweets']=="['@user']")
#df.drop(index[0],inplace=True)
X_train=[]
y_train=df['sentiment']
for i in df['processed_tweets']:
    X_train.append(' '.join(literal_eval(i)))
np.asarray(X_train)    
```

```python
dft= pd.read_csv('./ml-course-project-master/processed_testset.csv')
dft.head()
```

```python
X_test=[]
y_test=dft['sentiment']
for i in dft['processed_tweets']:
    X_test.append(' '.join(literal_eval(i)))
np.asarray(X_test)   
```

```python
# this is copied from MULTINOMIAL NOTEBOOK
def create_vocabulary(corpus):
    '''
    Creates a vocabulary out of the corpus.
    Input: List of strings, where each string is a processed tweet
    Output: List of words (vocabulary)
    '''
    # Initialize the vocabulary as an empty set
    vocab = {}
    for doc in corpus:
        # For every document, take the set union of the words in that document
        # with words already in the vocab
        words = doc.split(" ")
        vocab = set(vocab).union(words)
    return vocab
```

## USING TFIDF

```python
vocab = create_vocabulary(X_train)
vect=TfidfVectorizer(vocabulary=vocab)# converts words into vectors (NLP)
X_train_dtm=vect.fit_transform(X_train)
X_test_dtm=vect.transform(X_test)
```

```python
#svm with linear kernel
clf=svm.SVC(kernel='linear')
clf.fit(X_train_dtm,y_train)
y_predict= clf.predict(X_train_dtm)
print('train score:'+str(accuracy_score(y_train,y_predict)))
y_predict= clf.predict(X_test_dtm)
print('test score:'+str(accuracy_score(y_predict,y_test)))
```

```python
# Multinomial NB
clf=MultinomialNB()
clf.fit(X_train_dtm,y_train)
y_predict= clf.predict(X_train_dtm)
print('train score:'+str(accuracy_score(y_train,y_predict)))
y_predict= clf.predict(X_test_dtm)
print('test score:'+str(accuracy_score(y_predict,y_test)))
```

```python
# Logistic regression
clf=LogisticRegression()
clf.fit(X_train_dtm,y_train)
y_predict= clf.predict(X_train_dtm)
print('train score:'+str(accuracy_score(y_train,y_predict)))
y_predict= clf.predict(X_test_dtm)
print('test score:'+str(accuracy_score(y_predict,y_test)))
```

```python
#KNN with different k 1 to 10
from sklearn.neighbors import KNeighborsClassifier
for i in range(1,10):
    print('k='+str(i))
    clf=KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train_dtm,y_train)
    y_predict= clf.predict(X_train_dtm)
    print('train score:'+str(accuracy_score(y_train,y_predict)))
    y_predict= clf.predict(X_test_dtm)
    print('test score:'+str(accuracy_score(y_predict,y_test)))
```
